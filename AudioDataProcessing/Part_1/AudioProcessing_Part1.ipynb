```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
import librosa
```

```python
import numpy as np
samples = 100
f = 3
x = np.arange(samples)
y1=np.sin(2*np.pi*f*(x/samples))

plt.figure()
plt.stem(x,y1, 'r',)
plt.plot(x,y1)
plt.xlabel("Time-->")
plt.ylabel('<-amplitude->')
plt.show()
```

```python
import scipy
def fft_plot(audio, sampling_rate):
    n=len(audio)
    T=1/sampling_rate    
    yf=scipy.fft(audio)        
    xf=np.linspace(0.0, 1.0/(2.0*T), int(n/2))
    fig,ax = plt.subplots()
    ax.plot(xf,2.0/n*np.abs(yf[:n//2]))
    plt.grid()
    plt.xlabel("Frequency -->")
    plt.ylabel("Magnitude")
    return plt.show()
```

```python
fft_plot(y1,samples)
```
